# Human Capital: Education and Health Economics

## 1. Theoretical Framework: Human Capital and GDP
Human capital ($h$) is a primary driver of why some countries are richer than others, expanding on the standard production function: $Y=AF(K,hL)$. 

Development accounting isolates the impact of human capital on GDP through the following mathematical specifications:
* **Standard Specification:** $\frac{y_{1}}{y_{2}}=\frac{A_{1}k_{1}^{\alpha}h_{1}^{1-\alpha}}{A_{2}k_{2}^{\alpha}h_{2}^{1-\alpha}}=\frac{A_{1}}{A_{2}}\times(\frac{k_{1}}{k_{2}})^{\alpha}\times(\frac{h_{1}}{h_{2}})^{1-\alpha}$.
* **Capital-Output (K/Y) Specification:** $\frac{y_{1}}{y_{2}}=(\frac{A_{1}}{A_{2}})^{\frac{1}{1-\alpha}}\times(\frac{[K_{1}/Y_{1}]}{[K_{2}/Y_{2}]})^{\frac{\alpha}{1-\alpha}}\times(\frac{h_{1}}{h_{2}})$.

### Macroeconomic Mincer Regressions
To standardize the measurement of human capital across different school systems, macroeconomics incorporates the Mincerian return to schooling into the production function. 
* The function is specified as: $y=Ak^{\alpha}h^{(1-\alpha)} = Ak^{\alpha}e^{\theta s(1-\alpha)}$.
* Taking the natural logarithm yields the estimable macro regression: $\ln y = \ln A + \alpha \ln k + (1-\alpha)\theta s$.

### Measuring Educational Attainment
Barro and Lee (2013) compute average years of schooling using population shares and duration of educational levels.
* **Aggregate formula:** $s_{t}=\sum_{a=1}^{A}l_{t}^{a}s_{t}^{a}$, where $l_{t}^{a}$ is the population share of group $a$ in the population aged 25 and above, and $s_{t}^{a}$ is the number of years of schooling for that age group.
* **Age-group formula:** $s_{t}^{a}=\sum_{j}h_{j,t}^{a}Dur_{j,t}^{a}$, where $h_{j,t}^{a}$ is the fraction of group $a$ attaining educational level $j$ (primary, secondary, tertiary), and $Dur_{j,t}^{a}$ is the duration of that level in years.

### The Micro-Macro Paradox
There is an observable disconnect where microeconomic returns to schooling are estimated to be highly positive, yet macroeconomic returns (the boost in economic growth) often fail to materialize. 
* **Quantity vs. Quality:** Schooling does not strictly equal education; increasing enrollment without improving the quality of instruction yields lower returns.
* **Signaling:** If education merely signals inherent ability rather than building new productivity, individual wages rise but society's underlying productivity does not rise proportionally.
* **Skill Mismatch:** If labor demand constraints prevent the economy from absorbing skilled workers into high-productivity sectors, underemployment reduces aggregate productivity gains.

## 2. Education in the Developing World: Evidence from RCTs
While standard human capital models assume households optimize educational investments absent credit constraints, empirical evidence reveals time-inconsistent preferences and peer effects significantly alter these decisions. 

### Increasing School Quantity (Participation)
* **Reducing Costs and Subsidies:** Eliminating user fees and providing free uniforms can drastically reduce absenteeism; for example, free uniforms in Kenya reduced primary school absence by 64% among students who previously lacked them. 
* **Conditional Cash Transfers (CCTs):** Programs like PROGRESA in Mexico increase enrollment, particularly in secondary schools and for girls. Withholding part of a monthly CCT and distributing it when fees are due for the next school year yields higher subsequent enrollment than evenly spaced transfers, overcoming savings constraints and time-inconsistent preferences.
* **Information Provision:** Informing students and parents about the actual earnings differences between education levels can be a highly cost-effective way to boost attendance and reduce dropout rates, as seen in the Dominican Republic and Madagascar.
* **School-Based Health:** Mass deworming is one of the most cost-effective methods for increasing school participation, costing roughly $3.50 per additional year of education generated.

### Improving School Quality (Learning)
Simply providing more standard inputs (like textbooks or additional civil-service teachers) often fails to improve test scores because it does not address underlying systemic distortions like elite-oriented curricula and weak teacher incentives.
* **Curriculum Mismatch:** In centrally planned systems, curricula are often oriented toward elite students. Textbooks in rural Kenya failed to raise average test scores because they were written in English, which the median student could not read; they only benefited the initially highest-achieving quintile.
* **Pedagogical Reforms:** Interventions that allow for targeted instruction are highly effective. Tracking students by initial achievement, utilizing remedial tutoring programs (like the Balsakhi Program in India), and implementing computer-assisted learning successfully bypass the curriculum mismatch.
* **Teacher Incentives and Contracting:** Teacher absence rates are critically high in developing nations (e.g., 25% in India, 27% in Uganda). Monitoring attendance with cameras and paying teachers based on presence doubled attendance and raised test scores. Furthermore, hiring contract teachers locally—outside the protected civil-service system—saves money, lowers teacher absence, and raises student achievement.

## 3. The Economics of Poverty and Health
Poverty is strongly associated with poor health across and within countries, but the relationship is non-linear and exhibits varying impacts based on location and timing.

### Models of Health Production
* **The Preston Curve:** Demonstrates the relationship between life expectancy and GDP per capita. A standard empirical fit is $y = 6.2 \ln(x) + 13.8$ ($R^2 = 0.63$). The curve's upward shift over time implies that technological and medical improvements play a massive role independent of pure income growth.
* **Health Evolution Framework:** A baseline model of health without optimization dictates that health at age $a$ evolves as $H_a = H_{a-1} - \delta a^a + I + \epsilon_a$, where health deteriorates with age via the aging function $\delta a^a$, receives investments $I$, and absorbs stochastic shocks $\epsilon_a$. Since many health indicators capture "left-tail" events (like mortality thresholds), income has a massive effect on the poor (pushing them away from the threshold) but flat returns for the rich.

### Bidirectional Causality
1.  **Poverty Causes Poor Health:**
    * *In-Utero and Early Childhood:* Resources and maternal nutrition strictly impact birth outcomes and later-life adult health. In-utero exposure to stress, violence, and pollution leads to long-term detriments.
    * *Fetal Origins Hypothesis:* Poor early-life nutrition induces adaptations in metabolism or organ development that raise short-term survival but cause severe chronic problems later in life.
2.  **Poor Health Causes Poverty:**
    * Severe, sudden health events (like hospitalizations) result in persistent declines in employment, extensive earnings losses, and increased medical debt, often pushing households below the poverty line. 
    * *The Envelope Theorem:* Childhood health influences lifetime income by altering the productivity of human capital, without necessarily changing the volume of educational investment. Health improvements increase the speed of learning, elevating wages directly.

### 1. Please explain the difference between the gross and the net enrollment rate and explain how the average primary gross enrollment rate can be larger than 100%.

#### Gross vs. Net Enrollment Rates

The distinction between the Gross Enrollment Rate (GER) and the Net Enrollment Rate (NER) is fundamentally a difference in the formulation of their respective numerators. Both metrics normalize enrollment against the same demographic baseline (the denominator), but they capture distinct subsets of the enrolled population.

To formalize this, let us define the following parameters for a given educational system:
* $h$: A specific level of education (e.g., primary education).
* $a$: The officially designated age cohort for education level $h$ (e.g., ages 6 through 11).
* $P_{h,a}$: The total population of the official age cohort $a$.
* $E_h$: The total number of students enrolled in education level $h$, irrespective of age.
* $E_{h,a}$: The number of students enrolled in education level $h$ who strictly belong to the official age cohort $a$.
* $E_{h, \text{under}}$: The number of enrolled students who are younger than cohort $a$.
* $E_{h, \text{over}}$: The number of enrolled students who are older than cohort $a$.

From these definitions, total enrollment is the sum of the age-specific subgroups:
$$E_h = E_{h,a} + E_{h, \text{under}} + E_{h, \text{over}}$$

#### Net Enrollment Rate (NER)
The NER isolates the participation of the officially designated age cohort. It is the ratio of enrolled children in the official age group to the total population of that exact age group.

$$NER_h = \left( \frac{E_{h,a}}{P_{h,a}} \right) \times 100$$

**Properties:**
* Because the numerator $E_{h,a}$ is strictly bounded by the denominator $P_{h,a}$ (assuming accurate demographic tracking), the mathematical supremum of the NER is 100%.
* $NER_h \in [0, 100]$

#### Gross Enrollment Rate (GER)
The GER measures the total systemic load or capacity, capturing all enrolled students regardless of age, relative to the official target population.

$$GER_h = \left( \frac{E_h}{P_{h,a}} \right) \times 100$$

Substituting the expanded definition of $E_h$ yields:
$$GER_h = \left( \frac{E_{h,a} + E_{h, \text{under}} + E_{h, \text{over}}}{P_{h,a}} \right) \times 100$$

#### How Average Primary GER Can Exceed 100%
The mathematical condition for $GER_h > 100\%$ is simply $E_h > P_{h,a}$. 

Because the numerator includes students outside the designated age cohort ($E_{h, \text{under}} + E_{h, \text{over}}$), it is entirely possible for the total number of enrollees to surpass the total baseline population of the designated age cohort. 

$$E_{h,a} + E_{h, \text{under}} + E_{h, \text{over}} > P_{h,a}$$

##### Empirical Example
Assume a localized economy where the official primary education cohort is defined as ages 6 to 11.
* $P_{\text{primary}, 6-11} = 1,000$ (Total population of 6-11 year olds)
* $E_{\text{primary}, 6-11} = 850$ (Enrolled 6-11 year olds)
* $E_{\text{primary}, \text{over}} = 300$ (Enrolled students aged 12+, due to grade repetition or delayed entry)
* $E_{\text{primary}, \text{under}} = 0$

**Calculating NER:**
$$NER_{\text{primary}} = \left( \frac{850}{1000} \right) \times 100 = 85\%$$

**Calculating GER:**
$$GER_{\text{primary}} = \left( \frac{850 + 300 + 0}{1000} \right) \times 100 = \left( \frac{1150}{1000} \right) \times 100 = 115\%$$

##### Economic Drivers of GER > 100%
In development economics, a GER exceeding 100% typically identifies systemic inefficiencies or transitional shocks:
1.  **High Grade Repetition:** Standardized testing failures or poor instructional quality force students to repeat grades, thereby expanding the $E_{h, \text{over}}$ variable and keeping students in the primary system longer than the standard duration.
2.  **Delayed Entry:** Credit constraints or child labor dynamics may cause households to delay enrollment until a child is significantly older than the target starting age.
3.  **Policy Interventions:** The sudden abolition of user fees or the introduction of Conditional Cash Transfers (CCTs) often induces a massive backlog of previously un-enrolled, over-aged children to enter the primary system simultaneously to claim human capital investments.


### 2. Please explain (graphically) the Mincerian investment decision in education.

### The Mincerian Human Capital Investment Model

The Mincerian human capital model, pioneered by Jacob Mincer (1958, 1974), frames the decision to acquire education as a rigorous investment problem. An individual acts as a rational economic agent, weighing the present discounted value (PDV) of future earnings premiums against the immediate costs of schooling—both direct tuition costs and the opportunity cost of foregone earnings.



#### 1. The Mathematical Framework

Assume an individual is deciding how many years of schooling, $s$, to undertake before entering the labor market. 
* Let $w(s)$ be the constant annual wage received after completing $s$ years of schooling.
* Let $r$ be the individual's discount rate.
* Let $c$ be the direct, out-of-pocket cost of schooling per year (e.g., tuition).
* For mathematical elegance in the baseline model, assume an infinite working horizon.

The Present Value (PV) of earnings for an individual who chooses $0$ years of schooling and enters the labor market immediately is:
$$PV(0) = \int_{0}^{\infty} w(0) e^{-rt} dt = \frac{w(0)}{r}$$

The Present Value of earnings for an individual who chooses $s$ years of schooling involves paying cost $c$ for $s$ years, and then earning $w(s)$ from year $s$ onward:
$$PV(s) = \int_{0}^{s} -c \cdot e^{-rt} dt + \int_{s}^{\infty} w(s) e^{-rt} dt$$

Solving the integrals yields:
$$PV(s) = -\frac{c}{r}(1 - e^{-rs}) + \frac{w(s)}{r} e^{-rs}$$

#### The Equilibrium Condition
In a perfectly competitive labor market, the present value of all educational paths must be equal at the margin, meaning individuals are indifferent between $0$ and $s$ years of schooling because the market compensates them exactly for the delay and costs. Therefore, $PV(s) = PV(0)$:

$$\frac{w(0)}{r} = -\frac{c}{r}(1 - e^{-rs}) + \frac{w(s)}{r} e^{-rs}$$

Multiplying by $r$ and rearranging to solve for $w(s)$:
$$w(s) e^{-rs} = w(0) + c(1 - e^{-rs})$$
$$w(s) = w(0)e^{rs} + c(e^{rs} - 1)$$

If we assume public education where direct costs are zero ($c = 0$), the equation simplifies to the classic Mincerian earnings function. Taking the natural logarithm of both sides gives us the standard empirical specification:
$$\ln w(s) = \ln w(0) + rs$$

This proves that the percentage increase in earnings from an additional year of schooling must equal the discount rate $r$.

#### 2. The Graphical Anatomy

When we map this decision graphically (Age on the X-axis, Earnings on the Y-axis), we are visually comparing two distinct integral paths:

1.  **Path A (The High School Graduate):** The individual enters the labor force immediately at age 18. The earnings curve starts immediately at $w(0)$ and is relatively flat (ignoring age-experience profiles for the moment).
2.  **Path B (The College Graduate):** The individual goes to school for $s$ years. From age 18 to $18+s$, their "earnings" are negative, dropping below the x-axis to represent the direct costs ($-c$). At age $18+s$, they enter the labor market at a higher wage $w(s)$.

The visual model defines three critical areas:
* **Area 1: Foregone Earnings.** The income the student *could* have made if they were working instead of sitting in a classroom (the area under Path A, but above the x-axis, between years 0 and $s$).
* **Area 2: Direct Costs.** The out-of-pocket tuition expenses (the area below the x-axis, above Path B, between years 0 and $s$).
* **Area 3: The Gross Premium.** The increased wages earned over the rest of the working life (the area where Path B exceeds Path A).

For the investment in education to be economically rational, the **Present Discounted Value of Area 3** must be strictly greater than or equal to the **Present Discounted Value of Area 1 + Area 2**.



```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. Define Core Mathematical Parameters
# ==========================================
w0 = 30000          # Base annual wage for a high school graduate (w(0))
s = 4               # Years of schooling undertaken
r = 0.05            # Individual discount rate (5%)
c = 10000           # Direct annual cost of schooling (e.g., tuition)
age_start = 18      # Age of labor market/school entry
age_retire = 65     # Age of retirement

# Calculate the equilibrium wage after schooling w(s)
# This ensures PV(s) = PV(0) under continuous compounding
ws = w0 * np.exp(r * s) + c * (np.exp(r * s) - 1)

# ==========================================
# 2. Construct the Age and Earnings Arrays
# ==========================================
# Path A: High School Graduate (Enters labor market immediately)
age_A = np.linspace(age_start, age_retire, 500)
earn_A = np.full_like(age_A, w0)

# Path B: College Graduate (Schooling period + Working period)
# Period 1: In School (Ages 18 to 18+s)
age_schooling = np.linspace(age_start, age_start + s, 100)
earn_schooling = np.full_like(age_schooling, -c)

# Period 2: Working (Ages 18+s to 65)
age_working = np.linspace(age_start + s, age_retire, 400)
earn_working = np.full_like(age_working, ws)

# Combine Path B arrays for plotting a continuous line
age_B = np.concatenate([age_schooling, age_working])
earn_B = np.concatenate([earn_schooling, earn_working])

# ==========================================
# 3. Initialize the Plot
# ==========================================
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the baseline zero-axis
ax.axhline(0, color='black', linewidth=1.2, linestyle='--')

# Plot Path A (No Schooling)
ax.plot(age_A, earn_A, label=f'Path A: No Schooling ($w_0$ = ${w0:,})', 
        color='#1f77b4', linewidth=2.5)

# Plot Path B (With Schooling)
# We use a step function visual for the jump at graduation
ax.step(age_B, earn_B, where='post', label=f'Path B: {s} Years Schooling ($w_s$ = ${ws:,.0f})', 
        color='#2ca02c', linewidth=2.5)

# ==========================================
# 4. Fill the Integral Areas
# ==========================================
# Area 1: Foregone Earnings (Between w0 and 0 during schooling)
ax.fill_between(age_schooling, 0, w0, color='#1f77b4', alpha=0.2, 
                label='Area 1: Foregone Earnings')

# Area 2: Direct Costs (Between 0 and -c during schooling)
ax.fill_between(age_schooling, -c, 0, color='#d62728', alpha=0.2, 
                label='Area 2: Direct Costs')

# Area 3: Gross Premium (Between ws and w0 during working years)
ax.fill_between(age_working, w0, ws, color='#2ca02c', alpha=0.2, 
                label='Area 3: Gross Premium')

# ==========================================
# 5. Formatting and Annotations
# ==========================================
# Annotate the areas
ax.text(age_start + s/2, w0/2, 'Area 1', ha='center', va='center', fontsize=12, fontweight='bold', color='#0b5394')
ax.text(age_start + s/2, -c/2, 'Area 2', ha='center', va='center', fontsize=12, fontweight='bold', color='#990000')
ax.text((age_start + s + age_retire)/2, (ws + w0)/2, 'Area 3\n(Gross Premium)', 
        ha='center', va='center', fontsize=12, fontweight='bold', color='#274e13')

# Axis labels and title
ax.set_title('The Mincerian Human Capital Investment Decision', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Age', fontsize=14)
ax.set_ylabel('Annual Earnings / Costs ($)', fontsize=14)

# Set ticks and grid
ax.set_xticks(np.arange(age_start, age_retire + 5, 5))
ax.set_yticks(np.arange(-20000, ws + 10000, 10000))
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax.grid(alpha=0.3)

# Legend placement
ax.legend(loc='lower right', fontsize=11, framealpha=0.9)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

### 3. Please describe the way in which returns-to-education has been estimated in the macro literature and explain if there exists a micro-macro paradox and under which conditions the paradox is less likely to be present

### Estimating Returns to Education in the Macro Literature and the Micro-Macro Paradox

##### 1. Macroeconomic Estimation of Returns to Education

In macroeconomic literature, the estimation of returns to education begins by embedding human capital into an aggregate production function. The standard framework utilizes a Cobb-Douglas production function with Harrod-neutral (labor-augmenting) technological progress:

$$Y_t = K_t^\alpha (A_t h_t L_t)^{1-\alpha}$$

Where:
* $Y_t$ is aggregate output (GDP).
* $K_t$ is the physical capital stock.
* $A_t$ is total factor productivity (TFP) / technological knowledge.
* $L_t$ is the number of workers.
* $h_t$ is the average level of human capital per worker.
* $\alpha$ is the capital share of output.

To operationalize this for empirical estimation, macroeconomists rely on the **Mincerian human capital specification**. Assuming that human capital is generated through years of schooling ($s$) and that the return to a year of schooling is a constant $\theta$, the average human capital per worker is defined as:

$$h_t = e^{\theta s_t}$$

Substituting this into the production function and expressing it in per-worker terms ($y_t = Y_t/L_t$ and $k_t = K_t/L_t$) yields:

$$y_t = A_t k_t^\alpha (e^{\theta s_t})^{1-\alpha}$$

Taking the natural logarithm provides the foundational **Macroeconomic Mincer Regression** used in cross-country empirical studies:

$$\ln y_t = \ln A_t + \alpha \ln k_t + (1-\alpha)\theta s_t$$

To estimate the impact of education on economic *growth* (rather than just levels), researchers take the time derivative or first difference of the above equation. If we assume a constant steady-state capital-output ratio and constant technological growth, the growth rate of GDP per capita is modeled as a function of the change in average years of schooling:

$$\Delta \ln y_{i,t} = \gamma_0 + \gamma_1 \Delta \ln k_{i,t} + \beta \Delta s_{i,t} + \epsilon_{i,t}$$

Here, $\beta$ represents the macroeconomic return to an additional year of average schooling in country $i$. Average years of schooling ($s_{i,t}$) are typically derived from extensive demographic datasets, most notably the Barro and Lee (2013) educational attainment dataset, which aggregates enrollment and completion rates across age cohorts.

---

##### 2. The Micro-Macro Paradox

The **Micro-Macro Paradox** describes the stark empirical disconnect between microeconomic and macroeconomic estimates of the returns to education. 

* **The Microeconomic Finding:** Individual-level Mincer wage regressions ($\ln w_i = \beta_0 + \theta s_i + \text{controls}$) consistently yield high, positive, and statistically significant private returns to schooling. Across diverse economies, an additional year of schooling typically raises individual earnings by 7% to 10% ($\hat{\theta} \approx 0.07 - 0.10$).
* **The Macroeconomic Finding:** When researchers run the macroeconomic growth regression ($\Delta \ln y_{i,t} = \dots + \beta \Delta s_{i,t}$), the coefficient $\beta$ often turns out to be statistically insignificant, close to zero, or occasionally even negative. 

In short, while educating an individual dramatically increases their personal income, raising the *average* education level of a country’s population frequently fails to generate the proportional, measurable boost to aggregate GDP growth predicted by the $Y=AF(K,hL)$ model.

---

##### 3. Conditions Under Which the Paradox is Less Likely to be Present

The paradox is not absolute; it is an artifact of measurement, institutional quality, and labor market dynamics. The micro-macro disconnect diminishes (meaning macro returns align more closely with micro returns) under the following conditions:

##### A. When Education is Measured by Quality (Cognitive Skills) rather than Quantity
The most significant driver of the paradox is the assumption that $s_t$ (years sitting in a classroom) perfectly correlates with $h_t$ (actual productive knowledge). In many developing economies, systemic issues—such as severe teacher absenteeism, lack of targeted pedagogy, and elite-oriented curricula—mean that a year of schooling produces very little actual learning. 
* **Resolution Condition:** When macroeconomic models replace average years of schooling with measures of **educational quality** (such as standardized cross-country cognitive test scores in math and science, e.g., PISA or TIMSS), the paradox largely disappears. Growth regressions using cognitive skill aggregates show massive and highly significant macroeconomic returns, dwarfing the impact of mere enrollment duration.

##### B. When the Economy Has Sufficient Absorptive Capacity (Absence of Skill Mismatch)
Human capital only raises GDP if it is deployed in productive sectors. If a country rapidly expands higher education but its industrial base and technological frontier remain stagnant, the supply of skilled labor will outstrip demand.
* **Resolution Condition:** The paradox is less present in economies experiencing concurrent capital deepening ($K$) and technological adoption ($A$) that actively require skilled labor. In conditions where the labor market smoothly matches newly educated workers to high-productivity capital, the aggregate returns parameter ($\beta$) is positive and significant.

#### C. When Signaling and Social Exclusion are Minimized
If the labor market functions primarily on credentialism—where education acts as a sorting mechanism (Signaling Theory) or a rationing device to allocate a fixed number of primary-sector jobs (Social Exclusion Theory)—the micro return remains high (the individual gets the job) but the macro return is zero (no new aggregate value is created; it is a zero-sum displacement).
* **Resolution Condition:** The paradox is less likely in labor markets where compensation is strictly tied to marginal revenue product rather than monopoly rents or credential-based gatekeeping. When the acquired education imparts inherently productive skills (e.g., engineering, medical training) rather than just relative social rank, aggregate productivity rises alongside individual wages.

#### D. When Measurement Error in Panel Data is Corrected
From a purely econometric standpoint, the macro variable $\Delta s_{i,t}$ (the change in average schooling over a 5- or 10-year period) is highly susceptible to measurement error. Because average schooling evolves very slowly, much of the observed variance in $\Delta s_{i,t}$ across panel data is statistical noise, which biases the coefficient $\beta$ toward zero (attenuation bias).
* **Resolution Condition:** When researchers use high-reliability datasets and employ instrumental variables or long-difference estimators to filter out the classical measurement error in educational attainment data, the estimated macroeconomic returns rise significantly, converging toward the microeconomic estimates.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Set seed for reproducibility
np.random.seed(101)

# ==========================================
# 1. Simulate Microeconomic Data (Individual Level)
# ==========================================
n_individuals = 1000
# Individual years of schooling (normal distribution around 10 years)
schooling_i = np.random.normal(10, 2.5, n_individuals)
schooling_i = np.clip(schooling_i, 0, 20)

# True micro return to schooling is 10% (theta = 0.10)
theta = 0.10
# ln(w) = beta_0 + theta*s + error
log_wage_i = 8.0 + theta * schooling_i + np.random.normal(0, 0.4, n_individuals)

# Fit Micro OLS Regression
X_micro = sm.add_constant(schooling_i)
micro_model = sm.OLS(log_wage_i, X_micro).fit()
micro_pred = micro_model.predict(X_micro)

# ==========================================
# 2. Simulate Macroeconomic Data (Country Level)
# ==========================================
n_countries = 100
# Change in average years of schooling over a decade (Delta s)
delta_s_c = np.random.normal(1.5, 0.8, n_countries)

# True macro return is artificially close to 0 due to skill mismatch/credentialism
beta_macro = 0.01 
# Delta ln(y) = gamma_0 + beta*Delta_s + error (Growth rate of GDP per capita)
growth_y_c = 0.02 + beta_macro * delta_s_c + np.random.normal(0, 0.03, n_countries)

# Fit Macro OLS Regression
X_macro = sm.add_constant(delta_s_c)
macro_model = sm.OLS(growth_y_c, X_macro).fit()
macro_pred = macro_model.predict(X_macro)

# ==========================================
# 3. Visualization: The Micro-Macro Paradox
# ==========================================
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Plot 1: Microeconomic Returns ---
axes[0].scatter(schooling_i, log_wage_i, alpha=0.3, color='#1f77b4', edgecolor='w', s=40)
axes[0].plot(schooling_i, micro_pred, color='black', linewidth=3, 
             label=f'Micro Return ($\\hat{{\\theta}}$) = {micro_model.params[1]:.3f}')
axes[0].set_title('Microeconomics: Individual Wages vs. Schooling', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Individual Years of Schooling ($s_i$)', fontsize=12)
axes[0].set_ylabel('Log Annual Wage ($\ln w_i$)', fontsize=12)
axes[0].grid(alpha=0.3)
axes[0].legend(fontsize=12, loc='upper left')

# --- Plot 2: Macroeconomic Returns ---
axes[1].scatter(delta_s_c, growth_y_c, alpha=0.6, color='#d62728', edgecolor='w', s=60)
axes[1].plot(delta_s_c, macro_pred, color='black', linewidth=3, linestyle='--',
             label=f'Macro Return ($\\hat{{\\beta}}$) = {macro_model.params[1]:.3f}')
axes[1].set_title('Macroeconomics: GDP Growth vs. Average Schooling', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Change in Average Schooling ($\Delta s_t$)', fontsize=12)
axes[1].set_ylabel('GDP Per Capita Growth Rate ($\Delta \ln y_t$)', fontsize=12)
axes[1].grid(alpha=0.3)
axes[1].legend(fontsize=12, loc='upper left')

# Adjust layout and add a main title
plt.suptitle('Visualizing the Micro-Macro Paradox in Education Economics', fontsize=18, y=1.05, fontweight='bold')
plt.tight_layout()
plt.show()

# Print regression summaries for the notebook output
print("--- Microeconomic Regression Results ---")
print(f"Estimated private return to an additional year of schooling: {micro_model.params[1]:.2%}")
print(f"P-value: {micro_model.pvalues[1]:.4f} (Highly Significant)\n")

print("--- Macroeconomic Regression Results ---")
print(f"Estimated aggregate return to an average year of schooling: {macro_model.params[1]:.2%}")
print(f"P-value: {macro_model.pvalues[1]:.4f} (Statistically Insignificant)")
```

However, the geometric shapes of the data (the tight upward slope on the left versus the flat, noisy scatter on the right) are direct visual translations of the empirical findings from two highly distinct strands of economic literature. Economists rarely publish them side-by-side because microeconomists and macroeconomists traditionally publish in different journals using different methodologies. The paradox emerges when you place their respective literature next to each other.

Here is the foundational literature that dictates the exact shape of each graph:

### 1. The Microeconomic Visual (Left Panel)
The visual representation of individual log wages plotted against years of schooling is the most replicated graph in labor economics.
* **Jacob Mincer (1974), *Schooling, Experience, and Earnings*:** This is the genesis of the left panel. Mincer's original empirical work established that the relationship between log earnings and years of schooling is strikingly linear. 
* **George Psacharopoulos (1985, 1994, 2018):** Psacharopoulos's decades of global reviews on the returns to investment in education continuously validate this visual. His compilations of hundreds of micro-level wage regressions confirm that across almost all economies, the scatter plot of individual wages against schooling forms a tight, positive band with a slope (return) hovering around 7% to 10%. 

### 2. The Macroeconomic Visual (Right Panel)
The right panel—showing a flat or slightly negative relationship between the *change* in average schooling and economic *growth*—replicates the highly controversial scatter plots that disrupted macro-development economics in the late 1990s and early 2000s.
* **Lant Pritchett (2001), *"Where Has All the Education Gone?"* (World Bank Economic Review):** This is the definitive paper on the paradox. Pritchett plotted the growth rate of GDP per capita against the growth of human capital (using Barro-Lee data) for dozens of developing countries. His published scatter plots look exactly like the right panel of the simulation: a cloud of data points with a line of best fit that is flat or even slightly negative.
* **Benhabib and Spiegel (1994), *"The role of human capital in economic development evidence from aggregate cross-country data"*:** They ran standard cross-country growth regressions and repeatedly found that the coefficient on the change in human capital ($\Delta s$) was statistically insignificant, producing the "flat line" effect seen in the simulation.
* **Bils and Klenow (2000), *"Does Schooling Cause Growth?"* (American Economic Review):** They demonstrated that even when there is a positive correlation between schooling and growth, the causality likely runs in reverse (anticipated growth causes people to stay in school), meaning the strictly macroeconomic structural return to schooling is much flatter than previously assumed.

### Why Visualize Them Together?
The purpose of graphing them side-by-side in your notes is to instantly visualize the core tension in human capital theory: if the Mincerian literature (left) is undeniably true, how can the Pritchett literature (right) also be true? 

Visualizing the paradox this way forces the observer to rely on the theoretical bridging mechanisms—such as the Social Exclusion Theory (credentialism), skill mismatch, or the severe divergence between schooling quantity and educational quality—to explain how a perfectly rational microeconomic investment fails to scale into a macroeconomic growth engine.
